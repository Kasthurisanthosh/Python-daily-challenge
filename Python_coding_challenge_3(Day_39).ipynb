Project Title: Retail Inventory Manager

Scenario:
 You are managing an inventory system for a retail shop. Certain product categories are fixed, but the actual products can be added or removed by the staff.


```python
# Create a tuple with fixed product categories like Electronics, Stationery, and Clothing.
product_categories = ("Electronics","Stationery","Clothing")
print(f"product_categories: {product_categories}")
```

```python
# Create a list of products that can be updated dynamically.
products_list = ["Mobile","Laptop","Crayons","Kurtis"]
print(f"products_list: {products_list}")
```

```python
# Ask the user to enter a new product and add it to the list.
products_list.append(input("Enter a new product: "))
print(f"updated_product_list: {products_list}")
```

```python
# Ask the user to input the index of a product to remove it from the list.
print(f"Current Product List with Index: {list(enumerate(products_list))}")
removed_product=products_list.pop(int(input("Enter the index of the product to remove: ")))
print("The product ",removed_product,"was removed from the list")
```

```python
# Display the final list of products and the product categories.
print(f"Products_categories: {product_categories} ")
print(f"updated_product_list: {products_list}")
```
