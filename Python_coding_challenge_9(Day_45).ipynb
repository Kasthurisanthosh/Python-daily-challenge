Project Title: Simple Calculator App

Scenario:

 A company wants to provide users with a simple calculator to perform basic mathematical operations.


```python
# Create functions for addition, subtraction, multiplication, and division.
def add(a,b):
  return a+b

def sub(a,b):
  return a-b

def mul(a,b):
  return a*b

def div(a,b):
  if b==0:
    return "Cannot divide by zero"
  return a/b
```

```python
# Ask the user to input two numbers and the operation they want to perform.
print("Simple Calculator")
num1=float(input("Enter the first number: "))
num2=float(input("Enter the second number: "))

print("Choose an operation(Add(+),Sub(-),Mul(*),Div(/)):")
choice=input("Enter the operation: ")


if choice == '+':
    result = add(num1, num2)

elif choice == '-':
    result = sub(num1, num2)

elif choice == '*':
    result = mul(num1, num2)

elif choice == '/':
    result = div(num1, num2)
else:
    result = "Invalid choice! Please choose a valid option."

print("Result:", result)


# Use built-in functions like abs(), round(), and pow() within your solution.

print("Absolute value of result:", abs(result) if isinstance(result, (int, float)) else result)
print("Rounded result:", round(result, 2) if isinstance(result, (int, float)) else result)
print("Result raised to power 2:", pow(result, 2) if isinstance(result, (int, float)) else result)
```
