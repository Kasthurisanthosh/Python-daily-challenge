Project Title: Contact Directory

Scenario:

 You have been assigned to build a contact management system where employees can store and retrieve contact details efficiently.


```python
# Create a 2D list where each row holds a contact’s name, phone number, and email address.
contacts_list= [
    ["Kasthuri","+919876543210","kasthuri@gmail.com"],
    ["Santhosh","+919123456789","santhosh@gmail.com"],
    ["Athvika","+918019283764","athvika@gmail.com"]
]
print(f"\nContacts_List: {contacts_list}")
```

```python
# Allow the user to add a new contact.
new_contact_name=input("Enter the name of the contact: ")
new_contact_mobileno=input("Enter the mobile number of the contact: ")
new_contact_email=input("Enter the email of the contact: ")
contacts_list.append([new_contact_name,new_contact_mobileno,new_contact_email])
print(contacts_list)
```

```python
# Display all contacts sorted alphabetically by name.
contacts_list.sort()
print(contacts_list)
```

```python
# Allow the user to search for contacts by name. The search should support partial matches.
name=input("Enter your name: ").lower()
for i in contacts_list:
  if name in i[0].lower():
      print(i)
```

```python
# After sorting, display only the first 3 contacts in the list.
contacts_list.sort()
print(contacts_list[:3])
```
