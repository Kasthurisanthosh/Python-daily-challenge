Project Title: Student Grades & Hobbies Tracker

Scenario:

 A student activity coordinator is collecting information about student grades and hobbies. They want to ensure grades are updated properly and hobby lists are stored efficiently.


```python
# Create a dictionary with student names as keys and their grades as values.
student_grades = {"kasthuri":85,
            "santhosh":90,
            "athvika":95}
print(student_grades)
type(student_grades)
```

```python
# Allow the user to add or update grades for at least 3 students.
print("Add/Update grades for students")
for i in range(3):
  name = input(f"\nEnter student name {i+1}: ")
  grade = int(input(f"Enter grade for {name}: "))
  student_grades[name] = grade
print("\nUpdated Student Grades:")
print(student_grades)
```

```python
# Create a set of hobbies and allow the user to input at least 3 hobbies.
hobbies = set()
print("Enter 3 Hobbies")
for i in range(3):
  hobby = input(f"Enter hobby {i+1}: ")
  hobbies.add(hobby)
print("Hobbies set:",hobbies)
```

```python
# Convert the set of hobbies into a frozenset and display it
hobbies_frozen= frozenset(hobbies)
print("Hobbies set:",hobbies)
print("Hobbies frozenset:",hobbies_frozen)
```
