Day 38 – String Methods and Slicing

Project Title: Feedback Analyzer

Scenario:
 A content review team wants to analyse user feedback and comments. They want to check how frequently certain words appear and extract parts of the feedback for reporting.


```python
# Ask the user to input a sentence (representing feedback)
feedback = input("Please enter your feedback or comment: ")
```

```python
# Count how many words are in the sentence.
words = feedback.split()
num_words = len(words)
print("There are", num_words, "words in the feedback.")
```

```python
# Check how many times the word “data” appears in the sentence, regardless of case.
data_count=feedback.count('data')
print("The word 'data' appears", data_count, "times in the feedback.")
```

```python
# Extract and print the first 5 and last 5 characters of the feedback separately.
first_5_chars = feedback[:5]
last_5_chars = feedback[-5:]
print("The first 5 characters are:", first_5_chars)
print("The last 5 characters are:", last_5_chars)
```

```python
# Reverse the sentence and print it.
reversed_feedback = feedback[::-1]
print("The reversed feedback is:", reversed_feedback)
```
